```python
from pathlib import Path

from bs4 import BeautifulSoup
from loguru import logger

from squeezer import DATADIR, WORKDIR
```

```python
filename = Path("example.html")
html = DATADIR / filename

# Read
with html.open("r") as file:
    content = file.read()

# Build soup
logger.info("Parsing {}", html)
soup = BeautifulSoup(content, "html.parser")

# Prettify
pretty = Path(DATADIR / f"{filename.stem}_pretty.html")
logger.info("Beautify and save output to {}", pretty)   
with pretty.open("w") as file:
    file.write(soup.prettify())
```

```python
prods = soup.find_all("div", class_="product-list-box__content")

for prod in prods:
    prod_info = [x.strip() for x in str(prod).split("\n") if "<" not in x]
    print(",".join(prod_info))

```
